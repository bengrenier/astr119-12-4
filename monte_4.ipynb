```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 100000
xmin = 0
xmax = 1.75
ymin = -1
ymax = 1
```

```python
x = np.random.uniform(xmin,xmax,n)
y = np.random.uniform(-1,1,n)
```

```python
ir = np.where((y)<np.cos(x))[0]
ur = np.where((y)>=np.cos(x))[0]
```

```python
fig = plt.figure(figsize=(7,7))
plt.xlim([xmin-0.1,xmax+0.1])
plt.ylim([ymin-0.1,ymax+0.1])
plt.plot(x[ir],y[ir],'.',color="blue")
plt.plot(x[ur],y[ur],'.',color="0.75")
theta=np.linspace(xmin,xmax,n)
xc = theta
yc = np.cos(theta)
plt.plot(xc,yc,color="green")

plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()
```

```python
answer = np.sin(1.75) - np.sin(0)
print(answer)
```

```python
monte_approx = len(ir)/(float(n))
error_monte = (monte_approx-(answer))

print("Number of samples = ",n)
print("Approximate area = ",monte_approx)
print("Error in approx = ",error_monte)
```
